```python
 os.chdir('F:\\datasets')
import pandas as pd
import seaborn as sns
```

```python
tips = pd.read_csv('tip.csv')
titanic=pd.read_csv('titanic.csv')
```

# Scatter plot (Numerical-Numerical)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

# Bar Plot (Numerical - Categorical)

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

# Box Plot ( Numerical - Categorical)

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Fare'],hue=titanic['Survived'])
```

# DistPlot(Numerical-Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

# MarkMap ( Categorical-Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
(titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
#titanic.dtypes
```

# ClusterMap( Categorical-Categorical)

captures hierarchical relations in the data

```python
which values are closer to each other
```

```python
sns.clustermap(pd.crosstab(titanic['Sex'],titanic['Survived']))
```

# Pairplot

collection of scatter plot

```python
sns.pairplot(titanic,hue='Pclass')
```

# LinePlot(Numerical-Numerical)

join all the points of the scatter plot,lineplot should be used when the quantity on the x axis is time based 

```python
df = pd.read_excel('Test_set.xlsx', sheet_name='Sheet1')
df.head()
```

```python
sns.lineplot(x=df['Date_of_Journey'],y=df['Airline'])
```

```python
sns.clustermap(df.pivot_table(values='Total_Stops',index='Airline',columns='Date_of_Journey', aggfunc='count',fill_value=0))
```

```python
sns.heatmap(df.pivot_table(values='Total_Stops',index='Airline',columns='Date_of_Journey', aggfunc='count',fill_value=0))
```
